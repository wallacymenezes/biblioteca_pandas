```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas["estado"].fillna("MS", inplace=True)
vendas.info()
```

```python
media_precos = vendas["Preço"].mean()
vendas["Preço"].fillna(media_precos, inplace=True)
vendas.info()
```

```python
vendas[["Preço", "Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")
preco_correto = vendas.query("Preço <= Preço_com_frete")
```

```python
preco_correto.head()
```

```python
preco_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
preco_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(preco_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending = False).reset_index(),2)
```

```python
preco_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
preco_correto["mes"] = preco_correto["Data"].dt.month
preco_correto.head()
```

```python
clientes = pd.read_excel('/content/cliente_varejo.xlsx')
clientes.head()
```

```python
round(clientes[["idade", "renda"]].describe(),2)
```

```python
clientes.info()
```

```python
clientes = clientes.astype({"renda":"float"})
clientes.info()
```

```python
vendas_cliente = preco_correto.merge(clientes, how="left", on="cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").reset_index()
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
#sns.barplot(x="idcanalvenda", y="renda", data=agg_idcanal_renda)
plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="green")
plt.title("Média de Idade por Bandeira", loc="center", fontsize=14, fontweight="bold", color="black")
plt.xlabel("Bandeira")
plt.ylabel("Média de Idade")
plt.show()
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="green")
plt.title("Média de Renda por Canal de Venda", loc="center", fontsize=14, fontweight="bold", color="black")
plt.xlabel("Canal de Venda")
plt.ylabel("Média de Renda")
plt.show()
```

```python
venda_por_data = preco_correto.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(10,4))

#plt.plot(venda_por_data["Data"], venda_por_data["idcompra"], color="green")

plt.xlabel("Data da Venda")
plt.ylabel("Quantidade de Vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]

plt.plot(x, y, color="green", linewidth = "2")
plt.show()
```

```python
agg_dept_preco = round(preco_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending = False).reset_index(),2)
agg_dept_preco
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_dept_preco["Nome_Departamento"], agg_dept_preco["Preço_com_frete"], color="green")
plt.title("Média por Departamento", loc="center", fontsize=14, fontweight="bold", color="black")
plt.xlabel("Canal de Venda")
plt.ylabel("Preço com Frete")
plt.show()
```
